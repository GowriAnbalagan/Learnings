```python
#pandas is a library that contains functions used for datbase/tables and giving it a alias name as pd.
import pandas as pd
```

```python
#passing the input data file as a parameter to read_csv function and assigning it to varible dataset
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#printing the variable to see the data in the input file
dataset
```

```python
#Assigning a particular column (Year_of_experience) from dataset variable to independent variable
independent=dataset[["Year_of_experience"]]
independent
```

```python
#Assigning a particular column (Salary) from dataset variable to independent variable
dependent=dataset[["Salary"]]
dependent
```

```python
#sklearn is a library which supports for ML
#to save memory we are calling only the particular function in sklearn instead of importing all functions and consume high volume of memory
from sklearn.model_selection import train_test_split
#passing inputs/parameters to the function which splits the input and assigning those to corresponding variables
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
#Assigning LinearRegression class with non parameteried values to regresssor variable 
regressor=LinearRegression()
#calling fit function under Linear regreesion class to check/create the model and passing the training set data
regressor.fit(x_train,y_train)
```

```python
#Model is created and have to check the weight and bias as y=wx+b
#calling coefficient method and assigning it to variable
weight=regressor.coef_
weight
```

```python
#calling intercept variable and assigning it to variable bias/origin/intercept/min value
bias=regressor.intercept_
bias
```

```python
#calculating the predicted value by passing the test input 
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
#calculating the score by passing predicted value and actual value
r_score= r2_score(y_test,y_pred) 
r_score
```

```python
#saving the model using pickle library
import pickle
#.sav is the extension used in pickle to create a file
filename = "finalized_model_SLR.sav"
```

```python
#saving the model in created filename
pickle.dump(regressor,open(filename,'wb'))
```

```python
#checking the saved model is working fine by loading
loaded_model=pickle.load(open("finalized_model_SLR.sav",'rb'))
#passing a value to loaded model and predict the output
result=loaded_model.predict([[20]])
result
```
